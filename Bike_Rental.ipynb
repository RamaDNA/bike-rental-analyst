## Data loading

### Subtask:
Load the provided "hour.csv" file into a pandas DataFrame.


**Reasoning**:
Load the "hour.csv" file into a pandas DataFrame and display the first few rows to verify.



```python
import pandas as pd

try:
    df_hour = pd.read_csv('hour.csv')
    display(df_hour.head())
except FileNotFoundError:
    print("Error: 'hour.csv' not found. Please ensure the file exists in the current directory.")
    df_hour = None
except pd.errors.ParserError:
    print("Error: Could not parse 'hour.csv'. Please check the file format.")
    df_hour = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df_hour = None
```

## Data exploration

### Subtask:
Explore the loaded dataset to understand its characteristics.


**Reasoning**:
I need to explore the dataset by examining its shape, first few rows, last few rows, data types, descriptive statistics, missing values, distributions of key variables, and date range.



```python
# Display the shape of the DataFrame
print("Shape of the DataFrame:", df_hour.shape)

# Examine the first few rows
display(df_hour.head())

# Examine the last few rows
display(df_hour.tail())

# Get a summary of the DataFrame
display(df_hour.info())

# Generate descriptive statistics
display(df_hour.describe())

# Check for missing values
print("\nMissing values per column:\n", df_hour.isnull().sum())

# Key Variable Distributions and Date Range
import matplotlib.pyplot as plt

# Histograms for key variables
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.hist(df_hour['cnt'], bins=20, color='skyblue')
plt.title('Distribution of cnt')

plt.subplot(2, 3, 2)
plt.hist(df_hour['temp'], bins=20, color='salmon')
plt.title('Distribution of temp')

plt.subplot(2, 3, 3)
plt.hist(df_hour['hum'], bins=20, color='lightgreen')
plt.title('Distribution of hum')

plt.subplot(2, 3, 4)
plt.hist(df_hour['windspeed'], bins=20, color='gold')
plt.title('Distribution of windspeed')

plt.subplot(2, 3, 5)
plt.hist(df_hour['season'], bins=20, color='plum')
plt.title('Distribution of season')

plt.tight_layout()
plt.show()

# Date range
print("\nDate Range:")
print("Minimum Date:", df_hour['dteday'].min())
print("Maximum Date:", df_hour['dteday'].max())
```

## Data wrangling

### Subtask:
Convert the 'dteday' column to datetime objects and engineer new features from date and time information.


**Reasoning**:
Convert the 'dteday' column to datetime objects, extract new features (day of the week, month, hour of the day), and create additional features like 'is_weekend'.



```python
import pandas as pd

# Convert 'dteday' to datetime objects
df_hour['dteday'] = pd.to_datetime(df_hour['dteday'])

# Extract day of the week, month, and hour of the day
df_hour['day_of_week'] = df_hour['dteday'].dt.day_name()
df_hour['month'] = df_hour['dteday'].dt.month_name()
df_hour['hour_of_day'] = df_hour['hr']

# Create 'is_weekend' feature
df_hour['is_weekend'] = df_hour['day_of_week'].isin(['Saturday', 'Sunday']).astype(int)

# Verify data types
print(df_hour.dtypes)

# Display the head of the DataFrame
display(df_hour.head())
```

## Data analysis

### Subtask:
Analyze the relationship between bike rentals ('cnt') and other variables.


**Reasoning**:
Analyze the relationship between 'cnt' and other variables by calculating correlations and creating visualizations.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats

# Correlation analysis
numerical_features = ['temp', 'hum', 'windspeed', 'hour_of_day', 'is_weekend', 'cnt']
correlation_matrix = df_hour[numerical_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()


# Categorical variable analysis
categorical_features = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(categorical_features):
    plt.subplot(3, 3, i + 1)
    sns.barplot(x=feature, y='cnt', data=df_hour)
    plt.xticks(rotation=45, ha="right")
    plt.title(f'Bike Rentals by {feature}')
    plt.tight_layout()

plt.show()


# Time series analysis
plt.figure(figsize=(14, 6))
df_hour.groupby('dteday')['cnt'].sum().plot()
plt.xlabel('Date')
plt.ylabel('Total Bike Rentals')
plt.title('Total Bike Rentals Over Time')
plt.show()

plt.figure(figsize=(14, 6))
df_hour.groupby('day_of_week')['cnt'].mean().plot(kind='bar')
plt.xlabel('Day of the Week')
plt.ylabel('Average Bike Rentals')
plt.title('Average Bike Rentals by Day of Week')
plt.xticks(rotation=45, ha='right')
plt.show()

# Statistical tests (example: t-test for weekdays vs. weekends)
weekday_rentals = df_hour[df_hour['is_weekend'] == 0]['cnt']
weekend_rentals = df_hour[df_hour['is_weekend'] == 1]['cnt']

t_statistic, p_value = stats.ttest_ind(weekday_rentals, weekend_rentals)
print(f"T-test for weekday vs. weekend rentals: t-statistic={t_statistic:.2f}, p-value={p_value:.3f}")
```

## Data visualization

### Subtask:
Create visualizations to represent the key findings from the data analysis.


**Reasoning**:
Create visualizations based on the analysis results from the previous step. This includes time series plots, scatter plots, bar plots, and potentially combined visualizations.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Time Series of Bike Rentals
plt.figure(figsize=(14, 6))
df_hour.groupby('dteday')['cnt'].sum().plot()
plt.xlabel('Date')
plt.ylabel('Total Bike Rentals')
plt.title('Total Bike Rentals Over Time')
plt.show()

# 2. Relationship with Weather
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.regplot(x='temp', y='cnt', data=df_hour, color='skyblue')
plt.title('Bike Rentals vs. Temperature')

plt.subplot(1, 3, 2)
sns.regplot(x='hum', y='cnt', data=df_hour, color='salmon')
plt.title('Bike Rentals vs. Humidity')

plt.subplot(1, 3, 3)
sns.regplot(x='windspeed', y='cnt', data=df_hour, color='lightgreen')
plt.title('Bike Rentals vs. Windspeed')

plt.tight_layout()
plt.show()

# 3. Seasonality and Hour of Day Impact
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x='season', y='cnt', data=df_hour, palette="Blues_d")
plt.title('Average Bike Rentals per Season')

plt.subplot(1, 2, 2)
sns.barplot(x='hour_of_day', y='cnt', data=df_hour, palette="Greens_d")
plt.title('Average Bike Rentals per Hour')

plt.tight_layout()
plt.show()

# 4. Impact of Weekday/Weekend
plt.figure(figsize=(6, 5))
sns.barplot(x='is_weekend', y='cnt', data=df_hour, palette="magma")
plt.xticks([0, 1], ['Weekday', 'Weekend']) # Set custom x-axis labels
plt.title('Average Bike Rentals: Weekday vs. Weekend')
plt.show()
```
